```python
from langchain_openai import ChatOpenAI
from langchain.tools import tool

# Step 1: Define the tool
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b

# Step 2: Initialize the model
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# Step 3: Bind the tool
llm_with_tools = llm.bind_tools([multiply])

# Step 4: Invoke the model
result = llm_with_tools.invoke("Can you multiply 3 with 1000?")

# Step 5: Extract only the multiplication value
if hasattr(result, "tool_calls") and result.tool_calls:
    # Model decided to use your tool
    tool_call = result.tool_calls[0]
    output = multiply.invoke(tool_call["args"])  # âœ… actual numeric result
    print(output)  # ðŸ”¹ This prints just: 30
else:
    # Model gave an answer directly (no tool used)
    print(result.content)
```

```python
import requests
from langchain.tools import tool

@tool
def get_conversion_factor(base_currency: str, target_currency: str) -> float:
    """Get the live currency conversion rate between two currencies."""
    url = f'https://v6.exchangerate-api.com/v6/30a574917f314cff79ca7261/pair/{base_currency}/{target_currency}'
    response = requests.get(url)
    data = response.json()
    return data.get("conversion_rate", None)

@tool

def convert(base_currency_value: int, conversion_rate: float) -> float:
    """ Given a currency conversion rate this function calculates the target_currency value from a given phrase currency value"""
    return base_currency_value * conversion_rate

```
