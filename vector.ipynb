```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import Chroma
from langchain.schema import Document

docs = [
    Document(page_content="Virat Kohli is one of the most successful batsmen in IPL history",
             metadata={"team": "Royal Challengers Bangalore"}),
    Document(page_content="Sourav Gangully is a left handed player",
             metadata={"team": "KKR"}),
    Document(page_content="MS Dhoni, famous captain in Indian cricket team",
             metadata={"team": "Chennai Super Kings"})
]

vector_store = Chroma(collection_name="sports_db", embedding_function=embeddings)
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```python
vector_store.similarity_search(
  query='who among these are a bowler?',
  k=1
)
```

```python
vector_store.similarity_search_with_score(
  query='who among these are a bowler',
  k=1
)
```
